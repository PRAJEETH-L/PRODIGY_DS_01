Task 01
Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

For this task, I selected the Bank Data and created both a bar chart and a scatterplot.

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/data bank.csv')
data.head()
```

```python
year_to_visualize = '2019'
data_for_year = data[['Country Name', year_to_visualize]]
data_for_year = data_for_year.sort_values(by=year_to_visualize, ascending=False)

# Select the first 40 countries
data_for_year_top50 = data_for_year.head(50)

# Create a vertical bar chart using Seaborn
plt.figure(figsize=(12, 10))
sns.barplot(y='Country Name', x=year_to_visualize, data=data_for_year_top50, orient='h')
plt.title(f'Top 50 Countries - Distribution in {year_to_visualize}')
plt.xlabel('Country Name')
plt.ylabel('Population')
plt.show()
```

```python

year_1990 = '1990'
year_2000 = '2000'
data_selected_years = data[['Country Name', year_1990, year_2000]]

# Create a scatter plot using Seaborn
plt.figure(figsize=(10, 6))
sns.regplot(x=year_1990, y=year_2000, data=data_selected_years, line_kws={'color': 'red'})
plt.title(f'Population in {year_1990} vs. {year_2000}')
plt.xlabel(f'Population in {year_1990}')
plt.ylabel(f'Population in {year_2000}')
plt.show()
```
